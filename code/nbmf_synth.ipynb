# Batch VI for NBFM

```python
I = 1000
U = 2000
K = 50

r = 2.

np.random.seed(12345)
theta = np.random.gamma(0.1, 10, size=(K, U)).astype(np.float32)
beta = np.random.gamma(0.1, 10, size=(I, K)).astype(np.float32)
EX = beta.dot(theta)

lam = np.random.gamma(r, EX/r)
X = np.random.poisson(lam)
```

```python
reload(nbmf)
nbmf.DEFAULT_MIN_ITER = 10
coder = nbmf.NegBinomMF(n_components=K, max_iter=100, max_inner_iter=20, init_flat_lam=True, random_state=98765, verbose=True, r=1.)
```

```python
coder.fit(X)
```

```python
fig()
plt.subplot(211)
img(theta)
plt.colorbar()
plt.subplot(212)
img(coder.Et)
plt.colorbar()
```

```python
fig()
plt.subplot(211)
img(beta.T)
plt.colorbar()
plt.subplot(212)
img(coder.Eb.T)
plt.colorbar()
```

```python
fig()
plt.subplot(311)
img(X)
plt.colorbar()

plt.subplot(312)
img(coder.Elam)
plt.colorbar()

plt.subplot(313)
img(coder.Eb.dot(coder.Et))
plt.colorbar()
```

```python
plt.figure(figsize=(12, 3))
plt.subplot(121)
plt.hist(np.sqrt((coder.Et**2).sum(axis=0)), bins=50)
plt.subplot(122)
plt.hist(np.sqrt((coder.Eb**2).sum(axis=1)), bins=50)
plt.tight_layout()
pass
```
