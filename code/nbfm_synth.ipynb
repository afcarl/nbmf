```python
import functools

import numpy as np
import matplotlib.pyplot as plt

import nbmf
```

```python
fig = functools.partial(plt.figure, figsize=(10, 6))
img = functools.partial(plt.imshow, cmap='OrRd', aspect='auto', origin='lower', interpolation='none')
```

```python
I = 100
U = 200
K = 30

r = 2.

np.random.seed(12345)
theta = np.random.gamma(0.1, 10, size=(K, U)).astype(np.float32)
beta = np.random.gamma(0.1, 10, size=(I, K)).astype(np.float32)
EX = beta.dot(theta)

lam = np.random.gamma(r, EX/2)
X = np.random.poisson(lam)
```

```python
reload(nbmf)
coder = nbmf.NegBinomMF(n_components=K, max_iter=10, random_state=98765, verbose=True)
```

```python
coder.fit(X)
```

```python
fig()
plt.subplot(211)
img(theta)
plt.colorbar()
plt.subplot(212)
img(coder.Et)
plt.colorbar()
```

```python
fig()
plt.subplot(211)
img(beta)
plt.colorbar()
plt.subplot(212)
img(coder.Eb)
plt.colorbar()
```
